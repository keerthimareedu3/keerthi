3.T *attribute*

```python
import numpy as np
values=np.array([[12,23],[34,45]])
print("THE MATRIX:",)
print(values)
print("THE TRANPOSE MATRIX:",)
print(values.T)
print("The number of dimensions:",values.ndim)
print("The size of matix:",values.shape)
```

```python
import numpy as np
details=np.array([[23312,322323,34123,],[32143,45234,423445],[4312,3522,455432]])
print("THE MATRIX:")
print(details)
print("THE TRANPOSE MATIX:")
print(details.T)
print("The number of dimensions:",details.ndim)
print("The size of matix:",details.shape)
```
