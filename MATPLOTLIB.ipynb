# ***MATPLOTLIB ***

*1.Line chart*

```python
import matplotlib.pyplot as plt
x=[2,43]
y=[2,45]
plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[21,43,56]
y=[23,55,63]
z=[22,54,76]
plt.plot(x,y)
plt.plot(x,z)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[4,8,1]
y=[5,9,3]
plt.title("A line graph")
plt.xlabel('X-AXIS')
plt.ylabel('Y-AXIS')
plt.plot(x,y, marker='*')
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [5, 6, 9]
y = [5, 6, 20]
p = [10, 20, 25]

plt.title("A line graph")

plt.xlabel("X values")
plt.ylabel("Y values")

plt.plot(x, y, marker = '*')
plt.plot(x, p, marker = '*')
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[21,53,64,68]
y=[45,32,76,12]
z=[56,98,23,45]
plt.title("LINE GRAPH")
plt.xlabel("X-AXIS")
plt.ylabel("Y-AXIS")
plt.plot(x,y,marker='*')
plt.plot(x,z,marker='^')
plt.show()
```

*2.Bar chart*

```python
import matplotlib.pyplot as plt
months=["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
sales=[23,45,65,75,87,98,12,8,23,90,67,98]
plt.title('BAR GRAPH')
plt.xlabel("MONTHS")
plt.ylabel("SALES-AXIS")
plt.bar(months,sales)
plt.show()
```

```python
import matplotlib.pyplot as plt
product=["Laptop","moblie","T.V","Wash machine","Fridge"]
sales=[23,94,89,35,50]
plt.title("PRODUCT-SALES GRAPH")
plt.xlabel("PRODUCTS")
plt.ylabel("SALES")
plt.barh(product,sales)
plt.show()
```

```python
import matplotlib.pyplot as plt
items=["pizza","briyani","milkshake"]
sales=[324,5364,4315]
plt.title("SALES")
plt.xlabel("ITEMS")
plt.ylabel("SALES")
plt.bar(items,sales,width=1)
plt.show()
```

```python
import matplotlib.pyplot as plt
products=["laptop","moblie","fridge"]
profits=[678,342,987]
plt.title("PROFIT GRAPH")
plt.xlabel("PRODUCT")
plt.ylabel("PROFITS")
plt.bar(products,profits,width=1)
plt.show()
```
